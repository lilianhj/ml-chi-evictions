```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output_files/full_data_chicago.csv')
```

```python
numeric_df = df._get_numeric_data()
```

```python
numeric_df.columns
```

```python
cols = ['population', 'poverty-rate', 
        'pct-renter-occupied','median-gross-rent', 'median-household-income', 'median-property-value',
       'rent-burden', 'pct-white', 'pct-af-am', 'pct-hispanic',
       'pct-asian', 'avg_household_size_owner', 'avg_household_size_renter',
       'Total_Less_Than_HS', 'Total_HS_Grad',
       'Total_Bachelors', 'Mean_hours_worked']
```

```python
final_df = numeric_df[cols]
```

```python
final_df.replace(np.inf, np.nan, inplace=True)
final_df.replace(-np.inf, np.nan, inplace=True)
```

```python
final_df
```

```python
sns.scatterplot(x=cols[0], y=cols[1], data=final_df)
```

```python
sns.pairplot(final_df)
```

```python
final_df.hist(figsize=(20,20))
```
