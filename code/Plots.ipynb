## Plotting Histograms:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('output_files/test.csv')
numeric_data = data._get_numeric_data()

cols = ['population', 'poverty-rate',
       'renter-occupied-households', 'pct-renter-occupied',
       'rent-burden', 'pct-white', 'pct-af-am', 'pct-hispanic', 'pct-am-ind',
       'pct-asian','avg_household_size_owner',
       'median-gross-rent', 'median-household-income', 'median-property-value',
       'avg_household_size_renter', 'Total_Less_Than_HS', 'Total_HS_Grad',
       'Total_Some_College_or_AAS', 'Total_Bachelors', 'Mean_hours_worked',
       'crime_rate','if_top_10']

final_data = numeric_data[cols]

for col in final_data.columns[:-1]:
    final_data.groupby('if_top_10')[col].hist(alpha=0.4, figsize=(10,10))
    plt.title(col)
    plt.legend(['Negative', 'Positive'])
    plt.show()
```
